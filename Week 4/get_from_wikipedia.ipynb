```python
# import necessary libraries
import pandas as pd
import numpy as np
from bs4 import BeautifulSoup
import requests
```

```python
#get the URL 
website_url = requests.get('https://id.wikipedia.org/wiki/Kota_Bogor').text
```

```python
#print HTML 
soup = BeautifulSoup(website_url,'lxml')
print(soup.prettify())
```

```python
# find table 
My_table = soup.find('table',{'class':'wikitable sortable'})
My_table
```

```python
# get table data  from website 
links = My_table.findAll('a')
links
```

```python
#create table for data frame
Subdistric = []
for link in links:
    Subdistric.append(link.get('title'))
print(Subdistric)
```

```python
#import table to data frame 
import pandas as pd
df = pd.DataFrame()
df['Subdistric'] = Subdistric
print (df)
```

```python
#splitting the table into categories 
df[['Vilage','Subdistrics', 'City']] = df.Subdistric.str.split(",",expand=True,)
print(df)
```

```python
df_del_sub = df.drop('Subdistric', axis=1)
```

```python
print(df_del_sub)
```

```python
#Delete the City place in Subdistric 
df_rep = df_del_sub.replace(to_replace = "Bogor (halaman belum tersedia)", value ="Bogor") 
```

```python
print(df_rep)
```

```python
df_none = df_rep.drop(0)
```

```python
#save to CSV 
df_none.to_csv (r'C:\Users\BIrawan\Downloads\export_datalocation_city.csv', index = False, header=True)
```
